```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
# Remove the col1 imputed value
df1 = df.copy()

df1.iloc[1,0] = np.nan

df1


X = df1.iloc[[0,2,3,4],1:3]
y = df1.iloc[[0,2,3,4],0]
X
```

```python
lr = LinearRegression()
# Impute NaN values in X with the mean of each column before fitting
X_imputed = X.fillna(X.mean())
lr.fit(X_imputed,y)
# Predict using the original df1 row, as it doesn't contain NaNs in the relevant columns for prediction
print(lr.predict(df1.iloc[1,1:].values.reshape(1,2)))
```
