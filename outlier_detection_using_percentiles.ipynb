```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.head()
```

```python
df['Height'].describe()
```

```python
import seaborn as sns
sns.displot(df['Height'])
```

```python
sns.boxplot(df['Height'])
```

```python
upper_limit = df['Height'].quantile(0.99)
upper_limit
```

```python
lower_limit = df['Height'].quantile(0.01)
lower_limit
```

```python
new_df = df[(df['Height'] <= 74.78) & (df['Height'] >= 58.13)]
```

```python
new_df['Height'].describe()
```

```python
df['Height'].describe()
```

```python
sns.displot(new_df['Height'])
```

```python
sns.boxplot(new_df['Height'])
```

# caping


```python
np.where(new_df['Height']>= upper_limit,
         upper_limit,
         np.where(new_df['Height']<=lower_limit,
                  lower_limit,
                  new_df['Height']))
```
